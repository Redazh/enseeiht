## Interprétation 

1. Justifier les résultats obtenus pour l'exemple $f_0$ ci-dessus;
2. Justifier que l’algorithme implémenté converge en une itération pour $f_{1}$;
3. Justifier que l’algorithme puisse ne pas converger pour $f_{2}$ avec certains points initiaux.

**Remarque.** Vous pouvez mettre `affiche=true` dans les tests de l'algorithme de Newton pour
vous aider.


**1.Interpretation des résultats obtenus pour $f_0$.**

1. Pour le 1er cas, on pose `x0 = solution` comme point de départ,ce qui donne un gradient nul dès l'iteration 0.

2. Pour le 2eme cas, on fait `3` iterations afin de trouver une solution proche de `-pi/2` qui verifie la CN1.

3. Pour le 3eme cas, on pose `x0 = - solution` comme point de départ,ce qui donne un gradient nul dès l'iteration 0.


**2.Interpretation des résultats obtenus pour $f_1$.**

Pour $f_1$, comme nous l'avons vu en TD, il s'agit d'une fonction quadratique qui s'écrit sous cette forme : $f_1 = x^T H x + b^T + c$ . Et c'est un cas particulier ou l’algorithme converge en une itération.

**3.Interpretation des résultats obtenus pour $f_2$.**

L'algorithme puisse ne pas converger pour $f_2$ avec certains points initiaux si le point initial est loin de la solution et se trouve dans une zone ou la fonction presente des pentes abruptes.






```julia
using Plots
f2(x1, x2) = 100 * (x2 - x1^2)^2 + (1 - x1)^2
x1_range = -2:0.1:2
x2_range = -1:0.1:3
x1_grid = repeat(x1_range, 1, length(x2_range))
x2_grid = repeat(x2_range, 1, length(x1_range))'
f_values = f2.(x1_grid, x2_grid)
surface(x1_range, x2_range, f_values, xlabel="x1", ylabel="x2")
```
